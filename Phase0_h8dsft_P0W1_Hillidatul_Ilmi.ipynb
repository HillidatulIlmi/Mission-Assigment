# Query

- Mencari spesifikasi pada datasetnya

```python
mp.head()
```

```python
# pada dataframe akan di lakukan filtering
# melihat datafram dengan 1 kondisi
mp[
    mp['Total'] > 500
]
```

Menampilkan filtering kolom Total yang memiliki baris atas 500

```python
# Multiple dengan 4 kondisi
mp[
    (mp['Type 1']=='Water') &
    (mp['Total'] > 500) & 
    (mp['HP'] > 50) &
    (mp['Attack'] > 90)
]
```

```python
# Multiple dengan 4 kondisi
# Menampilkan kolom yang hanya di lakukan filtering saja
mp[
    (mp['Type 1']=='Water') &
    (mp['Total'] > 500) & 
    (mp['HP'] > 50) &
    (mp['Attack'] > 90)
][['Type 1', 'Total', 'HP', 'Attack']]
```

Hasil filtering pada dataframe di atas menampilkan kolom type 1 dengan baris water, kolom Total dengan nilai barisnya > 500, kolom HP dengan nilai barisnya > 50 dan kolom Attack dengn nilai barisnya > 90. Berdasarkan filtering menghasilkan data frame dengan 16 baris dan 4 kolom.

```python
# Multiple dengan 4 kondisi
# Menampilkan kolom yang hanya di lakukan filtering saja
mp[
    (mp['Type 1']=='Grass') & # memfilter kolom 1 dengan baris Grass
    (mp['Total'] >= 400) & # memfilter kolom total dengna baris nilai nya antara 400-700
    (mp['Total'] <= 700) & 
    (mp['HP'] % 2 !=0) & # memfiter kolom hp dengan baris berupa angka ganjil
    (mp['Attack'] > 90) #memfilter attack dengan baris di atas 90
][['Type 1', 'Total', 'HP', 'Attack']]
```

Hasil filtering pada dataframe di atas menampilkan kolom type 1 dengan baris Grass, kolom Total dengan nilai barisnya antara 400-700, kolom HP dengan barisnya bilangan ganjil dan kolom Attack dengan barisnya > 90. Berdasarkan filtering menghasilkan data frame dengan 16 baris dan 12 kolom.

```python
# Cara lain dengan memakai .query 
mp.query("Total >= 500 & HP <= 70 & Attack % 2 !=0")[['Total', 'HP', 'Attack']]
```

```python
mp.query("Total >= 500 & HP <= 70 & Attack % 2 !=0")[['Total', 'HP', 'Attack']].shape
#shape di gunakan untuk melihat jumlah baris dan kolom
```

```python
# penulisan cara lain pada .query
mp.query("Total >= 500 & \
        HP <= 70 & \
         Attack % 2 !=0").shape
```

Hasil filtering pada dataframe di atas menampilkan kolom Total dengan nilai barisnya antara >= 500, kolom HP dengan barisnya <= 70 dan kolom Attack dengan barisnya ganjil. Berdasarkan filtering dengan menggunakan perintah shape menghasilkan data frame dengan 27 baris dan 3 kolom.
